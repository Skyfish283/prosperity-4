```python
import pandas as pd
import matplotlib.pyplot as plt
alldata = pd.read_csv('kelp-1.csv',sep=';')
```

```python
kelp=alldata[alldata['product']=="KELP"]

plt.plot(kelp['timestamp'],kelp['mid_price'])
ax = plt.gca()
ax.set_xlim([0, 200000])
ax.set_ylim([2010, 2040])
plt.show()
```

```python
resin=alldata[alldata['product']=="RAINFOREST_RESIN"]

plt.plot(resin['timestamp'],resin['mid_price'])
ax = plt.gca()
ax.set_xlim([0, 200000])
ax.set_ylim([9990, 10010])
plt.show()
```
