```python
import pandas as pd

data = pd.read_csv('../data/round-1-island-data-bottle/prices_round_1_day_-1.csv',sep=';')
ink = data[data['product']=="SQUID_INK"]
change = []

for i in range(1,9999,1):
    curr = (ink.iloc[i]['ask_price_1'] + ink.iloc[i]['bid_price_1'])/2
    past = (ink.iloc[i-1]['ask_price_1'] + ink.iloc[i-1]['bid_price_1'])/2
    change.append(curr-past)
```

```python
import matplotlib.pyplot as plt
plt.plot(change)
```

```python
from statistics import *
pastup = []
pastdown = []
t = []
rsi = []

for i in range(0,9998,1):
    if change[i]>0:
        pastup.append(change[i])
    elif change[i] < 0:
        pastdown.append(-change[i])
    
    if pastup==[]:
        continue
    elif pastdown==[]:
        continue
    else:
        if len(pastup) > 14:
            del pastup[0]
        if len(pastdown) > 14:
            del pastdown[0]
        if mean(pastup)/mean(pastdown) == -1:
            continue
        rsi.append(100-(100/(1+mean(pastup)/mean(pastdown))))
        t.append(i*100)

plt.plot(t,rsi)

```
