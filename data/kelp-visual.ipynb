```python
import pandas as pd
import matplotlib.pyplot as plt
alldata = pd.read_csv('kelp-1.csv',sep=';')
```

```python
kelp=alldata[alldata['product']=="KELP"]

plt.plot(kelp['timestamp'],kelp['ask_price_1'])
ax = plt.gca()
ax.set_xlim([0, 200000])
ax.set_ylim([2010, 2040])
plt.show()
```

```python
resin=alldata[alldata['product']=="RAINFOREST_RESIN"]

plt.plot(resin['timestamp'],resin['mid_price'])
ax = plt.gca()
ax.set_xlim([0, 200000])
ax.set_ylim([9990, 10010])
plt.show()
```

```python
plt.plot(resin['timestamp'],resin['bid_price_1'])
plt.plot(resin['timestamp'],resin['ask_price_1'])
average_bid = []
average_ask = []
x = 1
sum = 0
aksum = 0
for i in range(len(resin)):
    if i==0:
        average_bid.append(resin['bid_price_1'].iloc[i])
        average_ask.append(resin['ask_price_1'].iloc[i])
        sum = resin['bid_price_1'].iloc[i]
        asksum = resin['ask_price_1'].iloc[i]
    else:
        sum+=resin['bid_price_1'].iloc[i]
        asksum += resin['ask_price_1'].iloc[i]
        average_bid.append(sum/(i+1))
        average_ask.append(asksum/(i+1))


plt.plot(resin['timestamp'],average_bid)
plt.plot(resin['timestamp'],average_ask)
ax = plt.gca()
ax.set_xlim([0, 200000])
ax.set_ylim([9990, 10010])
plt.show()
```

```python
running_avg_mid = []
running_avg_mid_sum = 0

for i in range(10):
    running_avg_mid_sum += resin['mid_price'].iloc[i]
    running_avg_mid.append(running_avg_mid_sum/(i+1))
running_avg_mid_sum /= 10
for i in range(10,2000,1):
    # print(running_avg_mid_sum, '+', (resin['mid_price'].iloc[i])/10,'-',(resin['mid_price'].iloc[i-10])/10)
    running_avg_mid_sum += (resin['mid_price'].iloc[i])/10
    running_avg_mid_sum -= (resin['mid_price'].iloc[i-10])/10
    # print(running_avg_mid_sum)
    running_avg_mid.append(running_avg_mid_sum)
# print(running_avg_mid)
plt.plot(running_avg_mid,resin['mid_price'])
# plt.plot(resin['timestamp'],resin['mid_price'])
```
