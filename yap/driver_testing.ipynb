```python

conv = [[1, 1.45, 0.52, 0.72], 
        [0.7,1,0.31,0.48],
        [1.95, 3.1, 1, 1.49],
        [1.34, 1.98, 0.64, 1]]



s = 2000000
t1 = []
t2 = []
t3 = []
t4 = []
t5 = []
t5d : Dict[int,List[int]] = {}
for i in range(4):
    t1.append(s*conv[3][i])
    for j in range(4):
        t2.append(t1[i]*conv[i][j])
        for k in range(4):
            t3.append(t2[i*4+j]*conv[j][k])
            for l in range(4):
                t4.append(t3[i*16+4*j+k] * conv[k][l])
                t5.append(t4[i*64+16*j+4*k+l]*conv[l][3])
                t5d[t4[i*64+16*j+4*k+l]*conv[l][3]] = [i,j,k,l]

print(t5d[max(t5)],max(t5))
max(t5)/2000000
```
